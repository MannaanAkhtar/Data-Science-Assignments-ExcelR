# Association Rules

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

# Apply the Apriori algorithm
frequent_itemsets = apriori(encoded_data, min_support=0.01, use_colnames=True)
frequent_itemsets

```python
# Generate the association rules
rules1 = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
rules1
```

```python
rules2 = rules1[["antecedents","consequents","support","confidence","lift"]]
rules2
```

```python
rules3 = rules2[(rules2["confidence"] >= 0.5) & (rules2["support"] >= 0.01) & (rules2["lift"] >= 1.5)]
rules3
```

### What is Lift and Why is it Important in Association Rules?
Lift is a key metric in Association Rule Mining that measures the strength of a rule by comparing the observed co-occurrence of items to their expected co-occurrence if they were independent. It helps determine whether the presence of one item increases or decreases the likelihood of another item appearing in a transaction.

###  What Are Some Limitations or Challenges of Association Rule Mining?
High Computational Cost:

The number of possible itemsets grows exponentially with the number of items, making it computationally expensive.
Many Uninteresting Rules:

Some rules may have high support and confidence but may not be useful in practice (e.g., trivial correlations).
Difficulty in Setting Thresholds:

Choosing appropriate thresholds for support, confidence, and lift can be challenging and domain-specific.
Rare Item Problem:

Items with low support may be ignored, even if they have strong associations.
Handling Continuous Data:

Association rule mining is better suited for categorical data; continuous data must be discretized, which can lead to loss of information.
Lack of Causality:

Association rules identify correlations, but they do not imply causation.
